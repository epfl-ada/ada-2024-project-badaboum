```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
```

```python
df_summaries = pd.read_csv('MovieSummaries/plot_summaries.txt', sep='\t', header=None, names=['Wikipedia_ID', 'Summary'])
df_summaries = df_summaries.head(1000)
```

```python
analyzer = SentimentIntensityAnalyzer()

df_summaries['Summary_Sentiment'] = df_summaries['Summary'].apply(lambda x : analyzer.polarity_scores(x))
```

```python
df_summaries['Summary_Sentiment_Compound'] = df_summaries['Summary_Sentiment'].apply(lambda x : x.get('compound'))
```

```python
df_summaries['Summary_Sentiment_Compound'].sample(10)
```

```python
plt.hist(df_summaries['Summary_Sentiment_Compound'], bins = 30)
```

```python
plt.boxplot(df_summaries['Summary_Sentiment_Compound'])
```
